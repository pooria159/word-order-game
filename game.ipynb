```python
import itertools
from tabulate import tabulate

def get_chars_input():
    chars_input = input("Enter characters separated by spaces: ")
    return chars_input.split()
def get_length_input():
    return int(input("Enter the length of the word to be generated: "))
def generate_permutations(chars_list, length):
    return [''.join(p) for p in itertools.permutations(chars_list, length)]

def main():
    chars_list = get_chars_input()
    length = get_length_input()
    permutations = generate_permutations(chars_list, length)
    table = [["Permutation"] + [f"Word {i+1}" for i in range(len(permutations))]]
    for i, word in enumerate(permutations):
        table.append([f"{i+1}"] + [word])
    print(tabulate(table, headers="firstrow", tablefmt="flex"))

if __name__ == "__main__":
    main()
```
